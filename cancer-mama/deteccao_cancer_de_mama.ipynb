# Importação das bibliotecas e dataset

```
#Importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import OneHotEncoder 
from sklearn.compose import ColumnTransformer 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, plot_confusion_matrix,accuracy_score, precision_score
```

```
#Fazendo a leitura dos dataset
df_data = pd.read_csv('D:/Users/artur/Projetos/ML/Dataset/data.csv')
```

```
#Teste para ver se o dataset foi importado corretamente
print(df_data.shape)
df_data.head()
```

# Análise Exploratória

```
# A função 'describe' nos trás algumas informações estatísticas básicas do nosso dataset para as colunas númericas
df_data.describe()
```

Podemos desconsiderar a feature 'id', pois é uma coluna apenas para identificação das amostras.

O dataset está completo. Sem nenhum dado missing, com exceção da última coluna não definida, onde todas as amostras são NaN, provavelmente foi uma coluna vazia no arquivo csv. A coluna será exluída logo abaixo. 

```
#Exlusão da coluna não definida
df_data.dropna(axis=1,inplace=True)
df_data.drop(columns='id',inplace=True)
```

```
# Grafico de correlação
diagnosis = {"M": 1, "B": 0}
df_data['diagnosis'] = df_data['diagnosis'].map(diagnosis)

df_data.corr().style.background_gradient().set_precision(2)
```

# Train an Test

```
#Separação da base treinameno e teste
X_train, X_test = train_test_split(df_data,test_size=0.3, random_state=42)
```

# Create Model

```
Y_train = np.array(X_train['diagnosis'])
X_train = np.array(X_train.drop('diagnosis',axis=1))
knn_model = KNeighborsClassifier()
knn_model.fit(X_train,Y_train)
```

# Predict

```
Y_test = np.array(X_test['diagnosis'])
X_test = np.array(X_test.drop('diagnosis',axis=1))
Y_predict = knn_model.predict(X_test);
```

```
plot_confusion_matrix(knn_model,X_test,Y_test); 
```

```
accuracy_score(Y_test,Y_predict)
```

```
precision_score(Y_test,Y_predict)
```
